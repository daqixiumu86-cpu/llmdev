```python
def _validate_str(username: str):
    if not 5 <= len(username) <= 15:
        print("エラー: ユーザー名は5文字以上15文字以内で入力してください")
        return False
    elif " " in username:
        print("エラー: ユーザー名に空白を含めることはできません")
        return False
    elif not username.isalnum():
        print("エラー: ユーザー名は英数字のみ使用できます")
        return False
    return True

username = str(input("ユーザー名を入力してください"))
if _validate_str(username):
    print("ユーザー名は有効です")
else:
    print("不正なユーザー名です")
```
