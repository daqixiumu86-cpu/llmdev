```python
import json
TODO_LIST_NAME = "todo_list.json"

try:
    with open(TODO_LIST_NAME, "r", encoding="utf-8") as file:
        todo_list = json.load(file)
except FileNotFoundError:
    todo_list = {}


def _get_next_available_key(todo_list: dict) -> int:
    used = {int(i) for i in todo_list.keys()}
    n = 1
    while n in used:
        n += 1
    return n

while True:
    try:
        operation_number = int(input("番号を入力してください[1: ToDoの追加, 2: ToDoの表示, 3:削除, 4: 終了・保存]:"))

        match operation_number:
            case 1:
                new_todo = input("追加するToDo項目を入力してください:")
                next_key = _get_next_available_key(todo_list)
                todo_list[str(next_key)] = new_todo
            case 2:
                print("ToDoリストを表示します:")
                for key, value in todo_list.items():
                    print(f"{key}: {value}")
            case 3:
                if not todo_list:
                    print("ToDoが空です。削除できません。")
                    continue

                for key, value in todo_list.items():
                    print(f"{key}: {value}")

                delete_num = input("削除したいToDoの番号を入力してください:")

                if delete_num in todo_list:
                    removed = todo_list.pop(str(delete_num))
                    print(f"削除しました: {delete_num}: {removed}")
                else:
                    print(f"入力値に一致するToDo項目が存在しません。入力値: {delete_num}")
            case 4:
                with open(TODO_LIST_NAME, "w", encoding="utf-8") as file:
                    json.dump(todo_list, file, ensure_ascii=False, indent=4)

                print("処理を終了します。")
                break
            case _:
                raise ValueError()
    except ValueError:
        print("不正な入力値です。")
        print("[1: ToDoの追加, 2: ToDoの表示, 3:削除, 4: 終了・保存]のいずれかを入力してください")
```
