```python
count_dict = {}
english_words = []

while True:
    english_word = input("英単語を入力してください:")

    if not english_word.strip():
        print(f"入力した英単語: {sorted(english_words)}")
        for i in sorted(count_dict):
            print(f"{i}が{count_dict[i]}個ありました")
        break
    else:
        english_words.append(english_word)
        for i in english_word:
            if i.isalpha():
                count_dict[i.lower()] = count_dict.get(i.lower(), 0) + 1
```
