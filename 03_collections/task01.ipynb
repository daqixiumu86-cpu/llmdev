```python
grades = [85, 90, 78, 92, 88]

average = sum(grades) / len(grades)
print(f"平均点: {average}")

sorted_grades = sorted(grades)
print(f"最高点: {sorted_grades[-1]}")
print(f"最低点: {sorted_grades[0]}")

above_averages = [i for i in grades if i >= average]
print(f"平均以上の学生数: {len(above_averages)}")

```
