```python
numbers = [1, 2, 3, 4, 5]
print(numbers)

fruits = ["apple", "banana", "cherry"]
print(fruits)

mixed = [1, "apple", 3.14, True]
print(mixed)

empty_list = []
print(empty_list)
```

```python
string_to_list = list("hello")
print(string_to_list)

tuple_to_list = list((1, 2, 3))
print(tuple_to_list)

range_to_list = list(range(5))
print(range_to_list)

empty_list = list()
print(empty_list)
```

```python
none_list = [None] * 5
print(none_list)

zeros = [0] * 10
print(zeros)
```

```python
original = [1, 2, 3]

copy1 = original[:]
print(copy1)

copy2 = list(original)
print(copy2)

copy3 = original.copy()
print(copy3)
```

```python
fruits = ["apple", "banana"]

fruits.append("cherry")
print(fruits)

fruits.insert(1, "orange")
print(fruits)

fruits.extend(["mango", "grape"])
print(fruits)
```
