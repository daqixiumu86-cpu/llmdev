```python
class BankAccount:
    def __init__(self, balance: int):
        self._balance = balance
    
    def get_balance(self):
        return self._balance
    
    def deposit(self, amount: int):
        if amount < 0:
            print("エラー: 正しい金額を入力してください")
            return False
        else:
            self._balance += amount
            return True
    
    def withdraw(self, ammount: int):
        if ammount < 0 or ammount > self._balance:
            print("エラー: 残高不足または無効な金額です")
            return False
        else:
            self._balance -= ammount
            return True

account = BankAccount(1000)
if account.deposit(500):
    print("500の入金に成功しました")

if account.withdraw(200):
    print("200の出金に成功しました")

print(f"現在の残高は {account.get_balance()} です") 

if account.withdraw(1500):
    print("1500の出金に成功しました")
```
