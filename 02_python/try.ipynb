```python
try:
    num = int(input("整数を入力してください: "))
    result = 100 / num
    print(f"100 を {num} で割った結果は {result} です")
except ZeroDivisionError:
    print("Error: 0で割ることはできません")
except ValueError:
    print("Error: 整数を入力してください")
except Exception as e:
    print(f"予期しないエラーが発生しました: {e}")
```

```python
a = 100
b = 0

try:
    if b == 0:
        raise ZeroDivisionError("0で割り算しようとしています")
    print(a/b)
except ZeroDivisionError as e:
    print(e)
```

```python
try:
    num = int(input("整数を入力してください: "))
    result = 100 / num
except ZeroDivisionError:
    print("エラー: 0 で割ることはできません")
except ValueError:
    print("エラー: 整数を入力してください")
else:
    print(f"計算結果: {result}")
finally:
    print("プログラム終了")
```
