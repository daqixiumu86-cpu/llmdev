```python
base_price = int(input("税抜き価格を入力してください。"))

import sys
if base_price < 1:
    print("税抜き価格が不正です。処理を終了します。")
    sys.exit()

include_tax_price = int(base_price * 1.1)
print(f"税込価格は{include_tax_price}円です。")

postage = 0
if include_tax_price < 2000:
    postage = 350
    print(f"送料は{postage}円です。")
else:
    print("送料は無料です。")

print(f"送料込税込価格は{include_tax_price + postage}です。")
```
