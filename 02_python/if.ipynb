```python
score = int(input("点数を入力してください: "))

if score >= 90:
    print("優秀です")
    print("大変よくできました！！")

elif score >= 60:
    print("合格です")
    print("おめでとうございます！")

else:
    print("不合格です")
    print("もっとがんばりましょう")

print("処理を終了します")
```

```python
score = 72  # 成績

match score // 10:
    case 10:
        print("満点です！")
    case 9 | 8:
        print("よくできました！")
    case 7 | 6:
        print("合格です！")
    case _:
        print("不合格です")
```
